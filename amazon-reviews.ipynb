```python
import pandas as pd
import numpy as np
```

```python
items = pd.read_csv('items.csv')
items.head()
```

```python
items.shape
```

```python
reviews = pd.read_csv('reviews.csv')
reviews.head()
```

```python
reviews.shape
```

```python
amazon = pd.merge(items, reviews, on = 'asin')
amazon.head(3)
```

```python
amazon.shape
```

```python
amazon.info()
```

```python
amazon = amazon[['asin', 'brand', 'title_x', 'title_y', 'totalReviews', 'rating_y', 'verified', 'body'  ]]
```

```python
amazon.dropna(inplace = True)
amazon.head()
```

```python
amazon.rename(columns = {'title_x':'product_title'}, inplace = True)
amazon.rename(columns = {'title_y':'product_review'}, inplace = True)
amazon.rename(columns = {'rating_y':'rating'}, inplace = True)
amazon.head()
```

```python
#convert to string
amazon['totalReviews'] = amazon['totalReviews'].apply(str)
amazon['rating'] = amazon['rating'].apply(str)
amazon['verified'] = amazon['verified'].apply(str)
amazon.head()
```

```python
amazon['tags'] =  amazon['product_review'] + " " + amazon['totalReviews'] + " " + amazon['rating'] + " " +  amazon['verified']
amazon.head()
```

```python
final_data = amazon[['asin', 'product_title', 'brand', 'tags']]
final_data.head()
```

```python
final_data['tags'] = final_data['tags'].apply(lambda x:x.lower())
```

```python
final_data.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    li = []
    for i in text.split():
        li.append(ps.stem(i))
        
    return " ".join(li)
```

```python
final_data['tags'] = final_data['tags'].apply(stem)
```

```python
final_data['tags']
```

```python
final_data.tags.str.split(expand=True).stack().value_counts()[:1000]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000,stop_words='english')
```

```python
vector = cv.fit_transform(final_data['tags']).toarray()
```

```python
vector[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(product):
    index = final_data[final_data['product_title'] == product].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(final_data.iloc[i[0]].product_title)
```

```python
recommend('Motorola I265 phone')
```
